# **All code wrapped in a class**

```python
class ProgrammingSolution:
  def __init__(self):
    pass

  def combineFiles(self,csv1,csv2,outputFile):

    #csv to pandas dataframe
    df_1=pd.read_csv(csv1,header=None)
    df_2=pd.read_csv(csv2,header=None)

    #dataframe to python dictionary such that the first columns becomes keys and others as value in a list
    df1_dict = df_1.set_index(df_1.columns[0]).T.to_dict('list')
    df2_dict = df_2.set_index(df_2.columns[0]).T.to_dict('list')

    result_dict={}

    #loop through each dictionary to match the keys
    for key1 in df1_dict:
      for key2 in df2_dict:
        if key1==key2:
          result_dict[key1]=df1_dict[key1]+df2_dict[key2]
        else:
          pass

    #write into the file output file
    with open(outputFile, "w", newline="") as file:
      writer = csv.writer(file)
      
      for key, values in result_dict.items():
          row = [key]
          row.extend(values)
          writer.writerow(row)

    return f'Combine file is written in the filepath : {outputFile}'
```

```python
path_for_csv1=os.path.join('/content/drive/MyDrive/smart_data_solutions/Java/resources/1.csv')
path_for_csv2=os.path.join('/content/drive/MyDrive/smart_data_solutions/Java/resources/2.csv')
output_file_path=os.path.join('/content/drive/MyDrive/smart_data_solutions/Java/outputFile.csv')

programmingSol = ProgrammingSolution()

programmingSol.combineFiles(path_for_csv1,path_for_csv2,output_file_path)
```

```python
#read output csv file as pandas dataframe 
df_result=pd.read_csv(output_file_path,header=None)
```
